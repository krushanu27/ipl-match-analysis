# üèè IPL Match Analysis (2008‚Äì2019)

This project explores IPL match data from 2008 to 2019 using Python and Pandas.

We'll analyze match-level trends like:
- Matches played each season
- Top teams by total wins
- How often toss winners also win the match

The goal is to extract insights and practice data cleaning, analysis, and visualization.


## üßπ 1. Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Data_Projects/matches.csv')

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df['season'] = df['season'].str.replace('IPL-', '')
df.head()
```

## üìä 2. Exploratory Data Analysis

### üìÜ Matches Per Season

We will count how many matches were played in each IPL season and visualize the trend.


```python
matches_per_season = df['season'].value_counts().sort_index()
matches_per_season.plot(kind='bar', color='cornflowerblue')
plt.title('Matches per Season (IPL)')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

### üèÜ Top 5 Teams by Wins

Let's find out which teams have won the most matches in IPL history.


```python
top_winners = df['winner'].value_counts().head(5).sort_values()
top_winners.plot(kind='barh', color='mediumseagreen')
plt.title('Top 5 IPL Teams by Total Wins')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.show()
```

### üé≤ Toss Winner vs Match Winner

We'll check how often the toss-winning team also won the match.


```python
toss_result = (df['toss_winner'] == df['winner']).map({True: 'Same Team', False: 'Different Team'})
toss_counts = toss_result.value_counts()

# Pie chart
toss_counts.plot(kind='pie', autopct='%.1f%%', colors=['green', 'red'])
plt.title('How Often the Toss Winner Also Won the Match?')
plt.ylabel('')
plt.show()
```

## üß† 3. Conclusion

- The IPL expanded over time, peaking around 2013.
- Mumbai Indians and CSK have the highest win totals.
- The toss does not strongly affect match outcomes ‚Äî only ~52% of toss winners won the match.
